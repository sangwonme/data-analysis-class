```python
import numpy as np
```

```python
arr = np.random.rand(6, 4)
arr
```

```python
# get dimension of array
arr.shape
```

```python
# selecting indecies of array
print(arr[:, 3])
print('--------------------------------------------------------------------')
print(arr[1, 1])
print('--------------------------------------------------------------------')
print(arr[0, :])
print('--------------------------------------------------------------------')
print(arr[:2, :3])
print('--------------------------------------------------------------------')
print(arr[[1, 3], 2])
```

```python
import pandas as pd

# Read file (txt file)
table = pd.read_csv('./data/sample_01.txt', delimiter='\t')
table = table.set_index('ID_REF')
table
```

```python
# Make np array from DataFrame

df_arr = np.array(table.loc['1007_s_at'])
print(df_arr)
print('shape : {}'.format(df_arr.shape))
print('--------------------------------------------------------------------')
df_arr = np.array(table.iloc[:2, :3])
print(df_arr)
print('shape : {}'.format(df_arr.shape))
```

```python
# Arithmetic Operations

arr1 = np.random.rand(2, 3)
arr2 = np.random.rand(2, 3)

print('ARR 1')
print(arr1, end='\n\n')
print('Arr 2')
print(arr2)
print('--------------------------------------------------------------------')
print(arr1 + arr2)
print('--------------------------------------------------------------------')
print(arr1 - arr2)
print('--------------------------------------------------------------------')
print(arr1 * arr2)
print('--------------------------------------------------------------------')
print(arr1 / arr2)
```

```python
# Operation : normalization

print(df_arr)
print('--------------------------------------------------------------------')
col_norm = np.linalg.norm(df_arr, axis=0).reshape(1,3)
print(col_norm)
cnorm_arr = df_arr / col_norm
print(cnorm_arr)
print('--------------------------------------------------------------------')
row_norm = np.linalg.norm(df_arr, axis=1).reshape(2,1)
print(row_norm)
rnorm_arr = df_arr / row_norm
print(rnorm_arr)
```
