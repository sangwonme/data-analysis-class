```python
import pandas as pd
```

```python
# Read file (txt file)
table = pd.read_csv('./data/sample_01.txt', delimiter='\t')
table
```

```python
# set first column as index
table = table.set_index('ID_REF')
table
```

```python
# select by value

print(table['GSM437093'])
print('-----------------------------------------------------------')
print(table.loc['1007_s_at'])
print('-----------------------------------------------------------')
print(table.loc['1007_s_at', 'GSM437093'])
print('-----------------------------------------------------------')
print(table.loc['1007_s_at':'121_at','GSM437093':'GSM437095'])
print('-----------------------------------------------------------')
print(table.loc[:,'GSM437093':'GSM437095'])
print('-----------------------------------------------------------')
print(table.loc[['1007_s_at', '1255_g_at'], ['GSM437093', 'GSM437097']])
```

```python
# select by index number

print(table.iloc[0])
print('-----------------------------------------------------------')
print(table.iloc[0, 3])
print('-----------------------------------------------------------')
print(table.iloc[:, 3])
print('-----------------------------------------------------------')
print(table.iloc[[3,5], [5, 7]])
```

```python
# Adding new row
new_row = pd.DataFrame([range(len(table.iloc[0]))], columns = table.columns, index=['NEW ROW'])
table = table.append(new_row)
table
```

```python
# Adding new col
new_col = pd.DataFrame(range(len(table)), columns = ['NEW COL'], index = table.index)
table['NEW COL'] = new_col
table
```

```python
# Deleting row/col
table = table.drop(['NEW ROW'])
table = table.drop(['NEW COL'], axis=1)
table
```
